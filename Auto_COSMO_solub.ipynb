```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import namedtuple
import json, os
import scipy.optimize
import ase
from ase.build import molecule
from ase.optimize import BFGS
from ase import Atoms
from ase.data.pubchem import pubchem_atoms_search, pubchem_atoms_conformer_search
#from gpaw import GPAW, PW
from ase.calculators.emt import EMT
from ase.io import write

plt.rc('font', size=16) # set font size in plots to 12
```

```python
# Input name or SMILES of compound

#no_salt = int(input('No of salts: '))
no_solvent = int(input('No of solvent: '))

print('Input method: name')

solvents = []
for c in range(no_solvent):
    solvent_input =input(f'Write name of solvent {c+1}: ')
    solvent_URL = solvent_input.replace(' ', '%20')
    solvents.append(solvent_URL)

#salts = []
#for c in range(no_salt):
#    salt_input = input(f'Write name of salt {c+1}: ')
#    salt_URL = salt_input.replace(' ', '%20')
#    salts.append(salt_URL)


for c in range(len(solvents)):
    solvent = pubchem_atoms_search(name=solvents[c])
    #print(solvent)
    print(solvent.positions)
    print(solvent.symbols)

    calc = EMT()
    solvent.calc = calc
    opt = BFGS(solvent, logfile='opt.log')
    opt.run(fmax=0.05)
    print(solvent.positions)
    print(solvent.symbols)
```
